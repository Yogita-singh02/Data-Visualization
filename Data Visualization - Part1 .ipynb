```python
df1 = sns.load_dataset('iris')
```

```python
df1
```

```python
df1['petal_width'].plot.hist()
```

```python
df1.plot(kind = 'hist' , stacked = True , bins = 50 , figsize = (10,10) , title = "histogram of iris")
```

```python
df1.hist()
```

```python
df1.hist(figsize= (10,10) , color = '#79ff4d' , alpha  = .9)
```

```python
df1.plot.scatter(x = 'sepal_width' , y = 'petal_width')
```

```python
df1.plot.scatter(x = 'sepal_length' , y = 'petal_width', c = 'petal_length')
```

```python
df1.plot.scatter(x  ='sepal_length' , y  = 'sepal_width', label = 'Arihant' ,s = 18 ,c = 'petal_width')
```

```python
df1.plot.hexbin(x  ='sepal_length' , y  = 'sepal_width', label = 'Arihant' ,gridsize = 10)
```
